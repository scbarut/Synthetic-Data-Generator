```python
parser = JsonOutputParser()

chain = prompt_template | llm | parser

subject = "Generate synthetic data for books in a library. columns must be atleast 9"
row_count = 3

try:
    result = chain.invoke({"text": subject, "row": row_count})
    print("Generated JSON Data:")
    print(result)

    dataset = pd.DataFrame(result)
except Exception as e:
    print(f"Error: {e}")
```
